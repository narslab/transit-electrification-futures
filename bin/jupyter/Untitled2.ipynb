```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8), dpi=300)

values = [32.37, 30.17]
colors = ['lightcoral', 'steelblue']
labels = ['CDB', 'HEB']

plt.bar(labels, values, color=colors)

plt.xlabel('Columns', fontsize=14)
plt.ylabel('Average VRIE (gal)', fontsize=14)

plt.show()
```
