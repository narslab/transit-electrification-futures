```python
import matplotlib.pyplot as plt
```

```python
# Charging cap data
years = list(range(2022, 2029))  # 2022 to 2028
caps = [15, 23, 23, 27, 38, 42, 52]

# Calculate the cap increase by 20% from 2029 to 2035
for year in range(2029, 2036):
    years.append(year)
    caps.append(caps[-1] * 1.2)  # increase by 20%

# Create a figure and a set of subplots
plt.figure(figsize=(10, 5), dpi=400)

# Plot the data
plt.plot(years, caps, marker='o')

# Title and labels
plt.title('Charging Capacity from 2022 to 2035', fontsize=20)
#plt.xlabel('Year')
plt.ylabel('Charging Cap (number of buses \n that can be charged daily)', fontsize=18)
plt.xticks(years, rotation=45, fontsize=16)  # Rotate the x labels for better readability
plt.yticks(fontsize=16)
# Show a grid
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
