```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
```

```python
# Re-load the CSV file due to the reset of the code execution state
#file_path = r'../../results/calibration-grid-search-BEB-oct2021-sep2022_12032023.csv'
file_path = r'../../results/calibration-grid-search-BEB-oct2021-sep2022_12062023.csv'
data = pd.read_csv(file_path)

# Selecting the relevant columns for the plot
parameters = ['gamma', 'driveline_efficiency_d_beb', 'battery_efficiency', 'motor_efficiency']
mape_train = data['MAPE_Energy_train']
mape_test = data['MAPE_Energy_test']

# Normalizing the fourth parameter for color mapping
norm = plt.Normalize(data[parameters[3]].min(), data[parameters[3]].max())
colors = plt.cm.viridis(norm(data[parameters[3]]))

# Adjusting the plot with increased font size and clearer labels
fig = plt.figure(figsize=(18, 8))

# Adjusting font size for labels and title
label_fontsize = 12
title_fontsize = 14

# Plot for MAPE Energy Train
ax1 = fig.add_subplot(121, projection='3d')
sc1 = ax1.scatter(data[parameters[0]], data[parameters[1]], data[parameters[2]], c=colors, marker='o')
ax1.set_title('MAPE Energy Train', fontsize=title_fontsize)
ax1.set_xlabel(parameters[0], fontsize=label_fontsize)
ax1.set_ylabel(parameters[1], fontsize=label_fontsize)
ax1.set_zlabel(parameters[2], fontsize=label_fontsize)

# Plot for MAPE Energy Test
ax2 = fig.add_subplot(122, projection='3d')
sc2 = ax2.scatter(data[parameters[0]], data[parameters[1]], data[parameters[2]], c=colors, marker='o')
ax2.set_title('MAPE Energy Test', fontsize=title_fontsize)
ax2.set_xlabel(parameters[0], fontsize=label_fontsize)
ax2.set_ylabel(parameters[1], fontsize=label_fontsize)
ax2.set_zlabel(parameters[2], fontsize=label_fontsize)

# Adding a color bar with label
cbar = plt.colorbar(sc1, ax=[ax1, ax2], shrink=0.5, aspect=10)
cbar.set_label(parameters[3], fontsize=label_fontsize)

plt.tight_layout()
plt.show()
```

```python
# Creating an improved plot for just the MAPE Energy Train with the color bar outside

fig = plt.figure(figsize=(10, 8))

# Plot for MAPE Energy Train
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(data[parameters[0]], data[parameters[1]], data[parameters[2]], c=colors, marker='o')
ax.set_title('MAPE Energy Train', fontsize=title_fontsize)
ax.set_xlabel(parameters[0], fontsize=label_fontsize)
ax.set_ylabel(parameters[1], fontsize=label_fontsize)
ax.set_zlabel(parameters[2], fontsize=label_fontsize)

# Adding a color bar with label, positioned outside the plot
cbar = plt.colorbar(sc, ax=ax, shrink=0.5, aspect=10, pad=0.1)
cbar.set_label(parameters[3], fontsize=label_fontsize)

plt.tight_layout()
plt.show()
```
