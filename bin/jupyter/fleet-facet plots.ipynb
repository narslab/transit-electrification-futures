```python
import matplotlib.pyplot as plt
from PIL import Image
import os

# List of image file paths in the order you want them to appear
image_files = [
    r'../../figures/l_FH_fleet.png',
    r'../../figures/m_FH_fleet.png',
    r'../../figures/h_FH_fleet.png',
    r'../../figures/l_BM_fleet.png',
    r'../../figures/m_BM_fleet.png',
    r'../../figures/h_BM_fleet.png',
    r'../../figures/l_FE_fleet.png',
    r'../../figures/m_FE_fleet.png',
    r'../../figures/h_FE_fleet.png',
]


# Define the column and row titles
columns = ['lo-cap', 'mid-cap', 'hi-cap']
rows = ['Hybrid', 'Hybrid Electric', 'Electric']

# Creating the figure and axes (3x3 grid in this case)
fig, axes = plt.subplots(3, 3, figsize=(15, 15), dpi=400)

# Looping through images, reading them, and plotting them on the respective axes
for i, row in enumerate(rows):
    for j, col in enumerate(columns):
        # Calculate the index for the current image
        img_index = i * len(columns) + j
        
        # Read and display the image
        img = Image.open(image_files[img_index])
        axes[i, j].imshow(img)
        axes[i, j].axis('off')  # Remove the axes
        
        # Setting the column titles for the top row
        if i == 0:
            axes[i, j].set_title(col, pad=20)  # Adjust 'pad' as necessary
        
        # Use fig.text to place the row labels at the desired positions
        if j == 0:
            # The x and y coordinates can be adjusted to place the labels correctly
            fig.text(0.04, (2-i)/3 + 0.1, row, va='center', ha='right', rotation='vertical', size='large')

plt.tight_layout()
plt.show()
```
