```python
top_10_speeds = SQ['Speed'].nlargest(200)
print(top_10_speeds)
```

```python
SQ=SQ.loc[SQ['Speed']<150]
```

```python
### 209 rows dropped with speed more than 150 mph
```

```python
SQ['Diesel'] = (SQ['Powertrain'].isin(['conventional', 'hybrid']) * SQ['Energy'])
```

```python
SQ['Diesel']
```

```python
result = SQ.groupby('Date')['Diesel'].sum().reset_index()
```

```python
result
```

```python
highest_energy_index = result['Diesel'].idxmax()
row_with_highest_diesel = result.loc[highest_energy_index]
```

```python
row_with_highest_diesel
```

```python
result = SQ.groupby('Date').agg({
    'TripKey': 'nunique',
    'Diesel': 'sum'
}).reset_index()

result.rename(columns={'TripKey': 'UniqueTripsCount'}, inplace=True)
```

```python
result
```

```python
highest_trips_index = result['UniqueTripsCount'].idxmax()
row_with_highest_trips = result.loc[highest_trips_index]
```

```python
row_with_highest_trips
```

```python
result.loc[highest_energy_index]
```

```python
reference_diesel = result[result['Date'] == '2021-10-29']['Diesel'].values[0]

# Compute the ratio for each day
result['DieselRatio'] = result['Diesel'] / reference_diesel
```
