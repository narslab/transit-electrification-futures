```python
# Merge the dataframes based on trip_id and route_id
merged_df = pd.merge(trips_df[['route_id', 'trip_id']], stop_times_df[['trip_id', 'stop_sequence']], on='trip_id')
merged_df = pd.merge(merged_df, routes_df[['route_id', 'route_short_name']], on='route_id')
```

```python
# Save to a CSV file with only the columns: trip-id, route, stop-sequence
final_df = merged_df[['trip_id', 'route_short_name', 'stop_sequence']]
final_df.columns = ['trip-id', 'route', 'stop-sequence']  # Rename columns to match the required format
final_df.to_csv('gtfs_output.csv', index=False)
```

```python
import pandas as pd

# Load the data into pandas DataFrames
trips_df = pd.read_csv(r'../../data/tidy/google_transit/trips.txt')
routes_df = pd.read_csv(r'../../data/tidy/google_transit/routes.txt')
stop_times_df = pd.read_csv(r'../../data/tidy/google_transit/stop_times.txt')

# Merge the dataframes based on trip_id and route_id
merged_df = pd.merge(trips_df[['route_id', 'trip_id']], stop_times_df[['trip_id', 'stop_sequence', 'stop_id']], on='trip_id')
merged_df = pd.merge(merged_df, routes_df[['route_id', 'route_short_name']], on='route_id')

# Save to a CSV file with the columns: trip-id, route, stop-sequence, stop-id
final_df = merged_df[['trip_id', 'route_short_name', 'stop_sequence', 'stop_id']]
final_df.columns = ['trip-id', 'route', 'stop-sequence', 'stop-id']  # Rename columns to match the required format
final_df.to_csv('gtfs_output.csv', index=False)
```
