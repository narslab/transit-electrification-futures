```python
import pandas as pd
```

```python
# Read CTF FY23 fuel tickets (June 2022 to September 2023)
df_valid_CTF_FY23 = pd.read_excel(r'../../data/tidy/CTF Fuel Tickets FY23 to date-tidy.xlsx')

# Convert date strings to datetime objects
df_valid_CTF_FY23['Transaction Date'] = pd.to_datetime(df_valid_CTF_FY23['Transaction Date'])

# Filter DataFrame to include transactions from July 2022 to end of September 2023
df_valid_CTF_FY23filtered = df_valid_CTF_FY23.loc[(df_valid_CTF_FY23['Transaction Date'] >= '2022-07-01') & (df_valid_CTF_FY23['Transaction Date'] <= '2022-09-30')]
df_valid_CTF_FY23filtered = df_valid_CTF_FY23filtered.drop(columns=["Unnamed: 0", "model"])
```

```python
# Read NTF FY23 fuel tickets (June 2022 to September 2023)
df_valid_NTF_FY23 = pd.read_excel(r'../../data/tidy/NTF Fuel Tickets FY23 to date-tidy.xlsx')

# Convert date strings to datetime objects
df_valid_NTF_FY23['Transaction Date'] = pd.to_datetime(df_valid_NTF_FY23['Transaction Date'])

# Filter DataFrame to include transactions from July 2022 to end of September 2023
df_valid_NTF_FY23filtered = df_valid_NTF_FY23.loc[(df_valid_NTF_FY23['Transaction Date'] >= '2022-07-01') & (df_valid_NTF_FY23['Transaction Date'] <= '2022-09-30')]
df_valid_NTF_FY23filtered = df_valid_NTF_FY23filtered.drop(columns=["Unnamed: 0"])
```

```python
# Read CTF FY22 fuel tickets (Oct 2021 to June 2022)
df_valid_CTF_FY22 = pd.read_csv(r'../../data/tidy/CTF-Fuel-Tickets-FY22-tidy.csv', delimiter=',', skiprows=0, low_memory=False)

# Convert date strings to datetime objects
df_valid_CTF_FY22['Transaction Date'] = pd.to_datetime(df_valid_CTF_FY22['Transaction Date'])

# Filter DataFrame to include transactions from Oct 2021 to June 2022
df_valid_CTF_FY22filtered = df_valid_CTF_FY22.loc[(df_valid_CTF_FY22['Transaction Date'] >= '2021-10-01') & (df_valid_CTF_FY22['Transaction Date'] <= '2022-06-30')]
df_valid_CTF_FY22filtered = df_valid_CTF_FY22filtered.drop(columns=['Unnamed: 0', 'model','vehicle_year', 'date'])
```

```python
# Read NTF FY22 fuel tickets (Oct 2021 to June 2022)
df_valid_NTF_FY22 = pd.read_csv(r'../../data/tidy/NTF-Fuel-Tickets-FY22-tidy.csv', delimiter=',', skiprows=0, low_memory=False)

# Convert date strings to datetime objects
df_valid_NTF_FY22['Transaction Date'] = pd.to_datetime(df_valid_NTF_FY22['Transaction Date'])

# Filter DataFrame to include transactions from Oct 2021 to June 2022
df_valid_NTF_FY22filtered = df_valid_NTF_FY22.loc[(df_valid_NTF_FY22['Transaction Date'] >= '2021-10-01') & (df_valid_NTF_FY22['Transaction Date'] <= '2022-06-30')]
df_valid_NTF_FY22filtered = df_valid_NTF_FY22filtered.drop(columns=['Unnamed: 0', 'model', 'vehicle_year', 'type', ])
```

```python
# Concatenate all four dataframes vertically
df_merged = pd.concat([df_valid_NTF_FY22filtered, df_valid_NTF_FY23filtered, df_valid_CTF_FY22filtered, df_valid_CTF_FY23filtered])

# Reset index
df_merged = df_merged.reset_index(drop=True)
```
