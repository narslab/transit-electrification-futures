```python
h_FH_variables_y = pd.read_csv(r'../../results/highcap-FH-optimized-variables-with-replacement-y.csv', low_memory=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by Year and Powertrain to sum up the Value
grouped_values = h_FH_variables_y.groupby(['Year', 'Powertrain'])['Value'].sum().unstack(fill_value=0)

# Reorder the columns
grouped_values = grouped_values[['CDB', 'HEB', 'BEB']]

# Plot
fig, ax = plt.subplots(figsize=(12, 12))

colors = {'CDB': 'lightcoral', 'HEB': 'steelblue', 'BEB': 'mediumseagreen'}
ax = grouped_values.plot(kind='bar', stacked=True, color=[colors[col] for col in grouped_values.columns], ax=ax)

# Annotate bars with the segment values for the last bar
for container_idx, container in enumerate(ax.containers):
    for rect_idx, rect in enumerate(container):
        if rect_idx == len(container) - 1:  # Annotate only the last bar
            height = rect.get_height()
            ax.annotate(f'{int(height)}',  # Dropped the decimals by converting to int
                        xy=(rect.get_x() + rect.get_width() / 2, rect.get_y() + height / 2),
                        ha='center', 
                        va='center',
                        color='black',  
                        fontsize=30)

# Set the title and labels with larger font sizes
ax.set_xlabel('Year', fontsize=34)
ax.set_ylabel('Fleet count', fontsize=34)

# Reverse the order of legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], fontsize=34)

ax.set_xticklabels(grouped_values.index, rotation=0)
ax.set_ylim(0, 230)

# Increase the font size for tick labels
ax.tick_params(axis='both', labelsize=28)

plt.tight_layout()

#fig.savefig(r'../../figures/h_FH_fleet.png', dpi=400)  

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to load data and add strategy and scenario
def load_data(file_path, strategy, scenario):
    data = pd.read_csv(file_path)
    data['Strategy'] = strategy
    data['Scenario'] = scenario
    return data

# Load all datasets with correct file paths
h_FH = load_data('../../results/highcap-FH-optimized-variables-with-replacement-y.csv', 'FH', 'high-cap')
m_FH = load_data('../../results/midcap-FH-optimized-variables-with-replacement-y.csv', 'FH', 'mid-cap')
l_FH = load_data('../../results/lowcap-FH-optimized-variables-with-replacement-y.csv', 'FH', 'low-cap')

h_BM = load_data('../../results/highcap-BM-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'BM', 'high-cap')
m_BM = load_data('../../results/midcap-BM-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'BM', 'mid-cap')
l_BM = load_data('../../results/lowcap-BM-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'BM', 'low-cap')

h_FE = load_data('../../results/highcap-FE-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'FE', 'high-cap')
m_FE = load_data('../../results/midcap-FE-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'FE', 'mid-cap')
l_FE = load_data('../../results/lowcap-FE-optimized-variables-with-replacement-40percentBEBcharging-y.csv', 'FE', 'low-cap')

# Combine all datasets
all_data = pd.concat([h_FH, m_FH, l_FH, h_BM, m_BM, l_BM, h_FE, m_FE, l_FE], ignore_index=True)

# Function to process and plot each dataset
def process_and_plot(data, strategy, scenario, ax):
    # Group by Year and Powertrain to sum up the Value
    grouped_values = data.groupby(['Year', 'Powertrain'])['Value'].sum().unstack(fill_value=0)

    # Reorder the columns
    grouped_values = grouped_values[['CDB', 'HEB', 'BEB']]

    # Plot
    colors = {'CDB': 'lightcoral', 'HEB': 'steelblue', 'BEB': 'mediumseagreen'}
    grouped_values.plot(kind='bar', stacked=True, color=[colors[col] for col in grouped_values.columns], ax=ax)


    # Set title, labels, and customizations
    ax.set_title(f'{strategy} - {scenario}', fontsize=34)
    ax.set_xlabel('Year', fontsize=34)
    ax.set_ylabel('Fleet count', fontsize=34)
    ax.set_xticklabels(grouped_values.index, rotation=0)
    ax.set_ylim(0, 250)
    ax.tick_params(axis='both', labelsize=28)

    # Reverse the order of legend
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], fontsize=34)

# Create a grid of plots
fig, axes = plt.subplots(3, 3, figsize=(36, 36))  # 3 strategies x 3 scenarios

# Plot each dataset in the grid
datasets = [(h_FH, 'FH', 'high-cap'), (m_FH, 'FH', 'mid-cap'), (l_FH, 'FH', 'low-cap'),
            (h_BM, 'BM', 'high-cap'), (m_BM, 'BM', 'mid-cap'), (l_BM, 'BM', 'low-cap'),
            (h_FE, 'FE', 'high-cap'), (m_FE, 'FE', 'mid-cap'), (l_FE, 'FE', 'low-cap')]

for (data, strategy, scenario), ax in zip(datasets, axes.flatten()):
    process_and_plot(data, strategy, scenario, ax)

plt.tight_layout()
plt.savefig('../../figures/facet_plot.png', dpi=400)
```

```python
# Function to process and plot each dataset
def process_and_plot(data, ax, show_xlabels, show_ylabels, show_legend):
    # Group by Year and Powertrain to sum up the Value
    grouped_values = data.groupby(['Year', 'Powertrain'])['Value'].sum().unstack(fill_value=0)

    # Reorder the columns
    grouped_values = grouped_values[['CDB', 'HEB', 'BEB']]

    # Plot
    colors = {'CDB': 'lightcoral', 'HEB': 'steelblue', 'BEB': 'mediumseagreen'}
    bars = grouped_values.plot(kind='bar', stacked=True, color=[colors[col] for col in grouped_values.columns], ax=ax, width=0.85, zorder=3)  # Adjusted width for wider bars and zorder for bars

    # Set labels and customizations
    if show_xlabels:
        ax.set_xlabel('', fontsize=34)
    else:
        ax.set_xlabel('')
        ax.set_xticklabels([])

    if show_ylabels:
        ax.set_ylabel('Fleet count', fontsize=34)
    else:
        ax.set_ylabel('')
        ax.set_yticklabels([])

    ax.set_ylim(0, 250)
    ax.tick_params(axis='both', labelsize=28)

    # Add horizontal grid lines behind the bars
    ax.grid(axis='y', linestyle='-', alpha=0.7, zorder=0)

    # Set more grid lines by specifying more y-ticks
    ax.yaxis.set_major_locator(plt.MultipleLocator(10))  # This sets the grid interval to 10

    # Add annotations for the last bar in each plot
    last_year = grouped_values.index[-1]
    total_last_year = grouped_values.loc[last_year].sum()
    
    # Calculate raw percentages
    raw_percentages = grouped_values.loc[last_year] / total_last_year * 100
    rounded_percentages = raw_percentages.round().astype(int)
    
    # Adjust the percentages to ensure they sum to 100
    diff = 100 - rounded_percentages.sum()
    if diff != 0:
        max_idx = rounded_percentages.idxmax()
        rounded_percentages[max_idx] += diff

    for col in grouped_values.columns:
        value = grouped_values.loc[last_year, col]
        percentage = rounded_percentages[col]
        if percentage > 0:
            bar_height = grouped_values.loc[last_year, :].cumsum()[col]
            ax.text(grouped_values.index.get_loc(last_year) + 0.05, bar_height - value / 2, f'{percentage}%',
                    ha='center', va='center', fontsize=20, color='white', fontweight='bold')  # Set color to white and fontweight to bold

    # Set legend
    if show_legend:
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles[::-1], labels[::-1], fontsize=34)
    else:
        ax.legend().set_visible(False)

# Create a grid of plots
fig, axes = plt.subplots(3, 3, figsize=(36, 36))  # 3 strategies x 3 scenarios

# Plot each dataset in the grid
datasets = [(l_FH, axes[0, 0]), (m_FH, axes[0, 1]), (h_FH, axes[0, 2]),
            (l_BM, axes[1, 0]), (m_BM, axes[1, 1]), (h_BM, axes[1, 2]),
            (l_FE, axes[2, 0]), (m_FE, axes[2, 1]), (h_FE, axes[2, 2])]

for idx, (data, ax) in enumerate(datasets):
    show_xlabels = idx >= 6  # Only show x-labels for the last row
    show_ylabels = idx % 3 == 0  # Only show y-labels for the first column
    show_legend = data is h_FH  # Show legend only for the top right plot
    process_and_plot(data, ax, show_xlabels, show_ylabels, show_legend)

# Set overarching titles (strategy names) on the right
row_titles = ['Hybrid', 'Hybrid+Electric', 'Electric']
for ax, row_title in zip(axes[:, -1], row_titles):  # Using the rightmost column
    ax.text(1.05, 0.5, row_title, fontsize=40, transform=ax.transAxes, ha='left', va='center', rotation=90)

# Set column titles
column_titles = ['lo-cap', 'mid-cap', 'hi-cap']
for ax, col_title in zip(axes[0], column_titles):
    ax.set_title(col_title, fontsize=40, pad=20)

plt.tight_layout()
plt.savefig('../../figures/facet-plot-annotated.png', dpi=600)
plt.show()
```
