```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import patches
import numpy as np
```

```python
df = pd.read_csv(r'../../results/computed-fuel-rates-oct2021-sep2022.csv', delimiter=',', skiprows=0, low_memory=False)
```

```python
df.columns
```

```python
# Copy required columns
df_clean = df[['ServiceDateTime','Route','Powertrain', 'Vehicle','VehicleModel','dist','Energy','Onboard','Date','Hour']].copy()
df_clean
```

```python
check = df_clean.nlargest(10, 'Energy')
check
```

```python
df_clean = df_clean.loc[df_clean['Energy']<300]
check = df_clean.nlargest(10, 'Energy')
check
```

```python
# Create a new column to show diesel consumption
df_clean['diesel'] = np.where(df_clean['Powertrain'] != 'electric', df_clean['Energy'], 0)
df_clean
```

```python
df_clean_filtered = df_clean.loc [df_clean['dist']!=0].copy()
df_clean_filtered['diesel-per-mile'] = df_clean_filtered['diesel'] /df_clean_filtered['dist'] 
```

```python
df_clean_filtered = df_clean_filtered.groupby(['Route','Date','Hour']).agg({'diesel-per-mile': ['sum'],'Onboard':['sum'], 'dist':['sum']}).reset_index()
df_clean_filtered.columns = df_clean_filtered.columns.droplevel(1)
df_clean_filtered
```

```python
df_clean_filtered = df_clean_filtered.groupby(['Route','Hour']).agg({'diesel-per-mile': ['mean'],'Onboard':['mean'], 'dist':['mean']}).reset_index()
df_clean_filtered.columns = df_clean_filtered.columns.droplevel(1)
df_clean_filtered
```

```python
df_clean_filtered['emissions-per-mile']=df_clean_filtered['diesel-per-mile']*0.01086879 ### For each gallon of diesel burned, 9.86 kgCO$_2$e of greenhouse gas (GHG) emissions are produced.
### 9.86kg = 0.01086879 us tons
```

```python
# Replace 'X' and '(X)' with an empty string
df_clean_filtered['Route'] = df_clean_filtered['Route'].str.replace('X', '', regex=False)
df_clean_filtered['Route'] = df_clean_filtered['Route'].str.replace(r'\(X\)', '', regex=True)

df_clean_filtered['Route'] = df_clean_filtered['Route'].str.replace(r'\(\)', '', regex=True)

# Remove any extra whitespace that may be left:
df_clean_filtered['Route'] = df_clean_filtered['Route'].str.strip()
```

```python
# Find the global maximum and minimum values for "emissions-per-mile"
global_max = df_clean_filtered['emissions-per-mile'].max()
global_min = df_clean_filtered['emissions-per-mile'].min()

# Create a list to store the new rows to be added
new_rows = []

# Iterate over each unique combination of Route and Hour
for route, hour in df_clean_filtered[['Route', 'Hour']].drop_duplicates().values:
    # Add a new row with the global maximum value for "emissions-per-mile"
    new_rows.append({'Route': route, 'Hour': hour, 'emissions-per-mile': global_max})
    # Add another new row with the global minimum value for "emissions-per-mile"
    new_rows.append({'Route': route, 'Hour': hour, 'emissions-per-mile': global_min})

# Convert the list of new rows to a DataFrame and append it to the original DataFrame
df_clean_filtered = df_clean_filtered.append(pd.DataFrame(new_rows), ignore_index=True)

# Display the first few rows of the extended DataFrame
df_clean_filtered.head()
```

```python
df_clean_filtered.to_csv(r'../../results/routes-emissions-for-video.csv', index=False)
```
