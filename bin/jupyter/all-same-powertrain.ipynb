```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_C = pd.read_csv(r'../../results/computed-fuel-rates-all-CDB.csv', low_memory=False)
```

```python
df_H = pd.read_csv(r'../../results/computed-fuel-rates-all-HEB.csv', low_memory=False)
```

```python
df_B = pd.read_csv(r'../../results/computed-fuel-rates-all-BEB.csv', low_memory=False)
```

```python
df_CDB=df_C.loc[df_C['Date']=='2021-10-29'].copy()
df_HEB=df_H.loc[df_H['Date']=='2021-10-29'].copy()
df_BEB=df_B.loc[df_B['Date']=='2021-10-29'].copy()
```

```python
# 1. Group each dataframe by TripKey and sum the Energy, and get the first VehicleModel value
grouped_CDB = df_CDB.groupby(['TripKey']).agg({'Energy': 'sum', 'VehicleModel': 'first'}).reset_index()
grouped_CDB.columns = ['TripKey', 'Energy_CDB', 'VehicleModel_CDB']

grouped_HEB = df_HEB.groupby(['TripKey']).agg({'Energy': 'sum', 'VehicleModel': 'first'}).reset_index()
grouped_HEB.columns = ['TripKey', 'Energy_HEB', 'VehicleModel_HEB']

# 2. Merge the grouped data of df_CDB and df_HEB on Tripkey
merged_data = pd.merge(grouped_CDB, grouped_HEB, on='TripKey', how='inner')

# 3. Compare the summed Energy values for each Tripkey between the two dataframes
condition = merged_data['Energy_CDB'] < merged_data['Energy_HEB']
filtered_data = merged_data[condition]

if not filtered_data.empty:
    print("There are trips where the Energy of df_CDB is smaller than the Energy of df_HEB:", len(filtered_data))
    print(filtered_data)
else:
    print("There are no trips where the Energy of df_CDB is smaller than the Energy of df_HEB.")
```
